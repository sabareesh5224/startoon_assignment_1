```
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()

filename = next(iter(uploaded))

with open(filename) as file:
    data = [float(line.strip()) for line in file.readlines()]

maxima = []
minima = []

for i in range(1, len(data) - 1):
    if data[i] > data[i-1] and data[i] > data[i+1]:
        maxima.append(i)
    elif data[i] < data[i-1] and data[i] < data[i+1]:
        minima.append(i)

plt.plot(data, label="Signal")
plt.scatter(maxima, [data[i] for i in maxima], color='red', label='Maxima')
plt.scatter(minima, [data[i] for i in minima], color='blue', label='Minima')
plt.legend()
plt.title("Signal with Maxima and Minima")
plt.xlabel("Index")
plt.ylabel("Amplitude")
plt.show()

print("Maxima indices:", maxima)
print("Minima indices:", minima)
```
